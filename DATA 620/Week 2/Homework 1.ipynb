#Introduction

For this homework, I chose to use the igraph library to draw out the social network. To use igraph, I went through the installation steps on their website. After that, I can simply import igraph by making the following call: 

```python
from igraph import *
```

#Drawing the Graph

Once we have the igraph library, I simply start a new graph by calling: 

```python
g = Graph()
```

After that, I can define the total number of vertices (i.e. people in this case) by the following line:

```python
g.add_vertices(10)
```

After that, I need to define the name of each of the vertices, which can be done via:

```python
g.vs["label"] = ["Carol", "Andre", "Fernando", "Diane", "Heather", "Ike", "Jane", "Beverly", "Garth", "Ed"]
```

With the above line, I have defined that the first edge is named as Carol, the second is Adre, and so on. After that, I can start definind the relationships within the graph itself. I'll start with the first edge, which is Carol:

```python
g.add_edges([
    (0,1),
    (0,2),
    (0,3)
])
```

For Carol, she is connected to Andrew, Fernando and Diane, which are 1,2 and 3 in terms of edge number, and Carol's edge number (id) is 0, so I simply include those relationships as pairs in a list as shown above. This is done for all the other people in the graph, leaving out already formed connections. For example, for Diane, we don't need to define connections for her, as defining connections from Andre, Carol, Fernando, Garth, Ed, and Beverly to her already takes care of all of her connections:

```python
# For Andre
g.add_edges([
    (1,2),
    (1, 7),
    (1,3)
])

# For Beverly
g.add_edges([
    (7,3),
    (7,8),
    (7,9)
])

# For Fernando
g.add_edges([
    (2,3),
    (2,8),
    (2,4)
])

# For Garth
g.add_edges([
    (8,4),
    (8,3),
    (8,9)
])

# For Ed
g.add_edges([
    (9,3)
])

# For Ike
g.add_edges([
    (5, 4)
])

# For Jane
g.add_edges([
    (6,5)
])
```

Once all the connections have been defined, I can simply plot the connection by calling the `plot` command as shown below:

```python
plot(g)
```
